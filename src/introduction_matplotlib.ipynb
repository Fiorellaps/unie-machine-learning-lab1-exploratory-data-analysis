## Abstract: Introduction to Matplotlib

[Matplotlib](https://matplotlib.org/) is the most widely used data visualization library in Python. It provides a MATLAB-like interface through its `pyplot` module, allowing you to create a wide range of static, animated, and interactive plots with just a few lines of code.

At its core, Matplotlib gives you full control over every element of a figure — from axes and labels to colors, markers, and layouts — making it the go-to tool for exploratory data analysis and publication-quality charts.

### Why Matplotlib?
- **Versatility:** Supports line plots, scatter plots, bar charts, histograms, heatmaps, and much more.
- **Integration:** Works seamlessly with NumPy, pandas, and the broader scientific Python ecosystem.
- **Customization:** Fine-grained control over every visual element of a plot.
- **Ecosystem:** Foundation for higher-level libraries like Seaborn and pandas plotting.

### Real-World Context

- **EDA:** Spotting trends, outliers, and distributions before modeling
- **Communication:** Presenting findings clearly to technical and non-technical audiences
- **Debugging:** Visually validating data transformations and model outputs
- **Reporting:** Generating charts for papers, dashboards, and presentations

## Sample Data

We'll use a simple dataset throughout this notebook to keep the focus on the plotting API rather than data wrangling.

````python
# create some data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
````

## Line Plot

The most basic plot type. Use `plt.plot()` to draw a line connecting data points. Always add a title and axis labels for clarity.

````python
# basic line plot
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

````python
# line plot with markers and legend
plt.plot(x, y, marker='o', linestyle='--', color='green', label='y values')
plt.title("Line Plot with Markers")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
````

## Scatter Plot

`plt.scatter()` plots individual data points without connecting lines. Useful for visualizing the relationship between two variables and spotting clusters or outliers.

````python
# scatter plot
plt.scatter(x, y, color='red')
plt.title("Scatter Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

## Bar Chart

`plt.bar()` creates vertical bars. Ideal for comparing discrete categories or groups.

````python
# bar chart
plt.bar(x, y, color='blue')
plt.title("Bar Chart")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

## Histogram

`plt.hist()` shows the distribution of a single variable by grouping values into bins. Essential for understanding data spread and skewness.

````python
import numpy as np

data = np.random.randn(1000)  # 1000 samples from a normal distribution
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
````

## Subplots

`plt.subplots()` lets you display multiple charts in a single figure using a grid layout. This is the recommended approach when comparing different views of the same data.

````python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(x, y, color='green', marker='o')
axes[0, 0].set_title("Line Plot")

axes[0, 1].scatter(x, y, color='red')
axes[0, 1].set_title("Scatter Plot")

axes[1, 0].bar(x, y, color='blue')
axes[1, 0].set_title("Bar Chart")

axes[1, 1].hist(np.random.randn(500), bins=20, color='purple', edgecolor='black')
axes[1, 1].set_title("Histogram")

plt.tight_layout()
plt.show()
````

## Customization

Matplotlib gives you fine-grained control over every visual element: grid lines, figure size, font sizes, tick marks, and more.

````python
plt.figure(figsize=(8, 4))
plt.plot(x, y, marker='s', linestyle='-', color='orange', linewidth=2, markersize=8, label='data')
plt.title("Customized Plot", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

## Pie Chart

`plt.pie()` displays proportions of a whole. Use `autopct` to show percentages and `startangle` to rotate the chart for better readability.

````python
labels  = ['Python', 'Java', 'JavaScript', 'C++', 'Other']
sizes   = [40, 20, 25, 10, 5]
explode = (0.1, 0, 0, 0, 0)  # slightly pull out the first slice

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%',
        startangle=90, colors=['steelblue','tomato','gold','mediumseagreen','plum'])
plt.title('Programming Language Popularity')
plt.show()
````

## Figure / Axes Object Model

Matplotlib has two interfaces:

- **`plt.` API** (stateful): quick and simple, good for single plots.
- **Object-oriented API** (recommended): explicitly work with `Figure` and `Axes` objects — more flexible and easier to manage in complex layouts.

| Concept | Description |
|---------|-------------|
| `Figure` | The entire canvas / window |
| `Axes`   | A single plot area within the figure |
| `Axis`   | The x or y axis of an Axes |

Always prefer the OO API when building anything beyond a single quick plot.

````python
# Create a Figure with one Axes
fig, ax = plt.subplots(figsize=(7, 4))

ax.plot(x, y, marker='o', color='steelblue', label='data')
ax.set_title('OO API — Line Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

````python
# Multiple Axes — each ax is independent
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.scatter(x, y, color='tomato')
ax1.set_title('Scatter')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.bar(x, y, color='gold')
ax2.set_title('Bar')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

plt.tight_layout()
plt.show()
````

## Styles

`plt.style.use()` applies a pre-built theme to all subsequent plots. This is the easiest way to improve the look of your charts without manual customization.

Some popular styles: `'ggplot'`, `'seaborn-v0_8'`, `'fivethirtyeight'`, `'dark_background'`, `'bmh'`.

````python
# List all available styles
print(plt.style.available)
````

````python
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(x, y, marker='o', label='y values')
ax.set_title('ggplot style')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.tight_layout()
plt.show()
````

````python
# Reset to default style
plt.style.use('default')
````

## Saving Figures

`plt.savefig()` exports the current figure to a file. Supported formats include PNG, PDF, SVG, and JPEG.

Key parameters:
- `dpi` — resolution (dots per inch); 150–300 for print quality
- `bbox_inches='tight'` — trims whitespace around the figure
- `transparent=True` — transparent background (useful for presentations)

````python
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(x, y, marker='o', color='steelblue', label='data')
ax.set_title('Saved Figure')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()

plt.savefig('my_plot.png', dpi=150, bbox_inches='tight')
print('Figure saved as my_plot.png')
plt.show()
````

## Reading and Displaying Images

Matplotlib can read and display images via `matplotlib.image.imread()`. An image is just a NumPy array of shape `(height, width, channels)` — so all the NumPy skills you've learned apply directly here.

In the cell below we download a sample image at runtime using `urllib` so no local file is needed.

````python
import urllib.request
import os
from matplotlib.image import imread

# Download a sample image into the local /images folder
os.makedirs("../images", exist_ok=True)
url      = "https://picsum.photos/300/200"
img_path = "../images/sample_image.jpg"
urllib.request.urlretrieve(url, img_path)
print("Image downloaded to", img_path)
````

````python
# Read the image into a NumPy array
image = imread(img_path)

print('type :', type(image))
print('shape:', image.shape)   # (height, width, channels)
print('dtype:', image.dtype)
````

````python
# Display the image
plt.figure(figsize=(6, 4))
plt.imshow(image)
plt.axis('off')  # hide axis ticks
plt.title('Sample Image')
plt.show()
````

````python
# Images are just NumPy arrays — slice and manipulate freely
red_channel   = image[:, :, 0]  # R
green_channel = image[:, :, 1]  # G
blue_channel  = image[:, :, 2]  # B

fig, axes = plt.subplots(1, 3, figsize=(12, 3))
for ax, channel, title, cmap in zip(
    axes,
    [red_channel, green_channel, blue_channel],
    ['Red', 'Green', 'Blue'],
    ['Reds', 'Greens', 'Blues']
):
    ax.imshow(channel, cmap=cmap)
    ax.set_title(title)
    ax.axis('off')

plt.tight_layout()
plt.show()
````
